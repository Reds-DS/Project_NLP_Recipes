```python
## Fonction qui calcule le nombre d'ingrédients -> Qui va nous permettre de rajouter une colonne
def count_ingredient(text_ingr):
    text = text_ingr.split(" ")
    return len(text)
```

```python
## Nettoyage de la colonne ingredient
ingredient_df["Ingredient"] = ingredient_df["Ingredient"].apply(lambda x : x.replace("_ ","_"))
ingredient_df["Ingredient"] = ingredient_df["Ingredient"].apply(lambda x : x.replace(" _","_"))
ingredient_df["Ingredient"] = ingredient_df["Ingredient"].apply(lambda x : x.replace("\\",""))
ingredient_df["Ingredient"] = ingredient_df["Ingredient"].apply(lambda x : x.lower())
```

```python
## Ajout de la colonne Nombre d'ingredients
ingredient_df["nbr_ingredient"] = ingredient_df["Ingredient"].apply(lambda text : count_ingredient(text))
ingredient_df.head()
```

```python
def classification(x):
    if x >=  12 :
        return 1
    elif x < 12 and x >= 10 : 
        return 2
    elif x < 10 and x > 5:
        return 3
    else:
        return 4
```

```python
ingredient_df["Classfication_nbr_ing"] = ingredient_df["nbr_ingredient"].apply(lambda x : classification(x) )
ingredient_df.head()
```

```python

## Cette fonction permet de trouver les ingrédients les moins fréquents dans notre colonne "Ingredient"

from collections import Counter
def freq_words(data_frame,colonne):
    text = ""
    word_freq = []
    low_freq_word = []
    L = [i for i in range(5)]
    for index,value in data_frame[[colonne]].iterrows():
        text += value[0] + " "
    word_list = text.split()
    cnt = Counter(word_list)
    for k, v in sorted(cnt.items(), key=lambda x: x[1], reverse=False):
        if v in L:
            low_freq_word.append(k)
    #return Counter(word_list).most_common()
    return low_freq_word
freq_words(ingredient_df,"Ingredient")
```

```python
## On enlève les mots moins fréquents pour optimiser notre prédiction
for word in freq_words(ingredient_df,"Ingredient"):
    ingredient_df["Ingredient"] = ingredient_df["Ingredient"].apply(lambda x : x.replace(word,"") if word in x else x)
```

```python
## Quelques classes utiles
# Classe pour sélectionner des colonnes à partir de leur nom
class SingleColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, key):
        self.key = key
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, data_dict):
        return data_dict[self.key]
```

```python
## Traitement du texte

import nltk
from nltk.corpus import stopwords

nlp = English()

def split_into_lemmas_spacy(desc):
    doc = nlp(desc)
    #return desc.split(" ")
    return [w.lemma_ for w in doc]

nltk_stopwords =stopwords.words('english')+list(string.punctuation)

## Transformer en sac de mots
desc_vectorizer = TfidfVectorizer(tokenizer = split_into_lemmas_spacy,
                                  lowercase = True,
                                  stop_words = nltk_stopwords,
                                  min_df = 0.001, ngram_range = (1,4))

## Pipeline 
desc_pipeline = make_pipeline(SingleColumnSelector(key = "Ingredient"),
                                                   desc_vectorizer)

desc_pipeline.fit(X_train)
resultat = desc_pipeline.transform(X_test.head())

print("Input")
print(X_test.Ingredient.head())
print()
print("Output")
print(resultat)
print()


feat_names = desc_vectorizer.get_feature_names()

print(feat_names)
```

```python

## Class qui permet de selectionner pluusieurs colonnes
class MultiColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        assert isinstance(X, pd.DataFrame)

        try:
            return X[self.columns].to_dict('records')
        except KeyError:
            cols_error = list(set(self.columns) - set(X.columns))
            raise KeyError("The DataFrame does not include the columns: %s" % cols_error)
```

```python
nbr_ingredient_vectorizer = DictVectorizer()

nbr_ingredient_pipeline = make_pipeline(MultiColumnSelector(columns = ["nbr_ingredient","Classfication_nbr_ing"]),
                                        nbr_ingredient_vectorizer)

nbr_ingredient_pipeline.fit(X_train)
resultat = nbr_ingredient_pipeline.transform(X_test.head())
print("Input")
print(X_test[["nbr_ingredient"]].head())
print()
print("Output")
print(resultat)
print()

## Trait extrait par la pipeline : "nbr_ingredient"
feat_names = nbr_ingredient_vectorizer.get_feature_names()

print(feat_names)
```

```python
union = FeatureUnion(transformer_list =[
                    ("nbr_ingredient_feature",nbr_ingredient_pipeline),
                    ("Ingredient_feature",desc_pipeline),
                    ])

preprocess_pipeline = make_pipeline(union)

preprocess_pipeline.fit(X_train)
X_transformed = preprocess_pipeline.transform(X_train)
print(X_transformed.shape)



fnames_description = desc_pipeline.named_steps['tfidfvectorizer'].get_feature_names()
fnames_nbr_ingredient = nbr_ingredient_pipeline.named_steps['dictvectorizer'].get_feature_names()

print('# description features : ', len(fnames_description))
print('# nbr_ingredient features: ', len(fnames_nbr_ingredient))
```

```python
fnames_description[10:50]
```

```python
## Entrainement de notre modèle 
## On a choisit la regression logistique pour classifier nos recettes



classifier_pipeline = make_pipeline(
    preprocess_pipeline,
    LogisticRegression(random_state = 7 , penalty = "l1" , max_iter = 200)
)
classifier_pipeline.fit(X_train,y_train)
predicted = classifier_pipeline.predict(X_test.head())
print(X_test.head())
print()
print("Predicted :")
print(predicted)
print()
print("Expected : ")
print(y_test.head())

```

```python
## predictions sur les données test
all_predictions = classifier_pipeline.predict(X_test)
print(all_predictions)
```

```python

## Matrix de confusion
print('accuracy', accuracy_score(y_test, all_predictions))
labels = np.unique(y_test)
cm =  confusion_matrix(y_test, all_predictions, labels=labels)
confusion_df = pd.DataFrame(cm, index=labels, columns=labels)
print('confusion matrix\n', confusion_df)
print('(row=expected, col=predicted)')
```

```python
print(classification_report(y_test, all_predictions))
```

```python
# Modèles à comparer
models = [
    ('Baseline', DummyClassifier(strategy='most_frequent')),
    ('Mutinomial NB', MultinomialNB()),
    ('CART', DecisionTreeClassifier()),
    ('LR', LogisticRegression(random_state = 7 , penalty = "l1" , max_iter = 200)),
    ('KNN', KNeighborsClassifier()),
    ('Random forest', RandomForestClassifier()),
    ('Gradient Boost', GradientBoostingClassifier())
]
# Evaluation de chaque résultat l'un après l'autre
scores = []
names = []
scoring = 'accuracy'

# Validation croisée à 5 plis
kfold = model_selection.KFold(n_splits=5, shuffle=True)
for name, model in models:
    names.append(name)
    model_pipeline = make_pipeline(preprocess_pipeline,model)
    y_pred = model_selection.cross_val_predict(model_pipeline, X, y, cv=kfold)
    print(name)
    print(classification_report(y, y_pred))
    acc = metrics.accuracy_score(y, y_pred)
    scores.append(acc)
    
# Représentation graphique
indices = np.arange(len(scores))
fig = plt.figure()
plt.barh(indices, scores, .2, label="score", color='b')
plt.yticks(())
for i, c in zip(indices, names):
    plt.text(-.3, i, c)
plt.show()
```
